```python
#Try and Except statement is used to handle these errors within our code in Python.
#The try block is used to check some code for errors 
#i.e the code inside the try block will execute when there is no error in the program.
#Whereas the code inside the except block will execute whenever the program encounters some error in the preceding try block.
import ast
def convert(obj):
    L=[]
    try:
        for i in ast.literal_eval(obj):
            L.append(i['name'])
    except:
        L.append('none')
    return L
```

```python
test_data['Keywords']=test_data['Keywords'].apply(convert)
test_data['spoken_languages']=test_data['spoken_languages'].apply(convert)
test_data['genres']=test_data['genres'].apply(convert)
test_data['belongs_to_collection']=test_data['belongs_to_collection'].apply(convert)
train_data['Keywords']=train_data['Keywords'].apply(convert)
train_data['spoken_languages']=train_data['spoken_languages'].apply(convert)
train_data['genres']=train_data['genres'].apply(convert)   
train_data['belongs_to_collection']=train_data['belongs_to_collection'].apply(convert)   
```

```python
def convert3(obj):
    L=[]
    counter=0
    try:
        for i in ast.literal_eval(obj):
            if counter<3:
                L.append(i['name'])
                counter+=1
            else:
                break
    except:
        L.append('none')
    return L
    
```

```python
test_data['cast']=test_data['cast'].apply(convert3)
train_data['cast']=train_data['cast'].apply(convert3)
```

```python
test_data.head(1)
```

```python
def fetch_director(obj):
    L=[]
    try:
        for i in ast.literal_eval(obj):
            if i['job']=='Director':
                L.append(i['name']) 
                break
    except:
        L.append('none')
    return L
```

```python
test_data['crew']=test_data['crew'].apply(fetch_director)
train_data['crew']=train_data['crew'].apply(fetch_director)
```
